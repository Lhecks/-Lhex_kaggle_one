```python
def point_request(key_api):
    # Get the first hour of today
    start = arrow.now().floor('day')

    # Get last hour of today
    end = arrow.now().ceil('day')

    responses = requests.get(
        'https://api.stormglass.io/v2/weather/point',
        params={
            'lat': 58.7984,
            'lng': 15.257015,
            'params': ','.join(['airTemperature', 'currentSpeed', 'humidity', 'waveHeight']),
            'start': start.to('UTC').timestamp(),  # Converting to UTC timestamp
            'end': end.to('UTC').timestamp()  # Converting to UTC timestamp
        },
        headers={
            'Authorization': key_api
        }
    )

    json_datas = responses.json()
    print(json_datas)
    print('\njson Data:\n')

    for data in json_datas['hours']:
        air_temp = data['airTemperature']
        times = data['time']
        humidity = data['humidity']
        print('Air Temperature: {}'.upper().format(air_temp))
        print('Time: {}'.upper().format(times))
        print('humidity: {}'.upper().format(humidity))
        print()
```
